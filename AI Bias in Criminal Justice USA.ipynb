```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
def convert_to_numerical(df, column):
  categorical_features = df[column]
  le.fit(categorical_features)
  numerical_features = le.transform(categorical_features)
  df.insert(df.columns.get_loc(column) + 1, column = (column + '_categorical'), value = numerical_features)

convert_to_numerical(df, 'State')
```

```python
X = df[['State_categorical', 'Black-White Dissimilarity Index', 'Total # of Arrests (2013-2018)', 'Total Population (2021)', 'Budget Per Capita for Police', 'Total # of Educated']]
Y = df[['Total # of People Killed (2013-2021)']]

from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.33, random_state=42)

yy_test = Y_test.to_numpy()
```

```python
from sklearn import tree


clf = tree.DecisionTreeRegressor(min_samples_split = 6, max_depth = 4, max_leaf_nodes=4)

clf = clf.fit(X_train, Y_train)
prediction_test = clf.predict(X_test)

Y_predictions = clf.predict(X_test)

clf.score(X_test, Y_test)

np.sqrt(((yy_test -Y_predictions)** 2).mean())
```

```python
from sklearn.metrics import mean_squared_error
error = mean_squared_error(Y_test, prediction_test)

(np.sqrt(error)/df['Total # of People Killed (2013-2021)'].mean())

from sklearn.metrics import mean_absolute_error
error2 = mean_absolute_error(Y_test, prediction_test)

((error2)/df['Total # of People Killed (2013-2021)'].mean())
```

```python
from sklearn.metrics import r2_score
r2_score1 = r2_score(Y_test, prediction_test)

r2_score1
```
